## Filtering Data from the Data Frame

```python
import pandas as pd
```

```python
# Data set import
bios = pd.read_excel('./data/olympics-data.xlsx')
bios.info()
```

```python
# Filtering the data
# Get the name and height for players taller than 215cm, both are the same but diff syntax
bios.loc[bios['height_cm'] > 215, ['name', 'height_cm']]
bios[bios['height_cm' ] > 215][['name', 'height_cm']]

# Multiple conditions, height above 215 and born country USA
bios.loc[(bios['height_cm'] > 215) & (bios['born_country']=='USA')]

# Seeing if a string contains something, then if so output name and home country
bios.loc[bios['name'].str.contains("Keith"), ['name', 'born_country']]

# Keith OR Patrick
bios.loc[bios['name'].str.contains("Keith|Patrick"), ['name', 'born_country']]

# Keith OR Patrick, output in decending order from date died
bios.loc[bios['name'].str.contains("Keith|Patrick"), 
         ['name', 'died_date', 'born_country']].sort_values('died_date', ascending=False)

# Records with one of these countries, then see if we have a first name Jamess
# Then output name, born country, and birthday (oldest to youngest)
bios.loc[bios['born_country'].isin(["USA", "FRA", "GBR"]) 
         & (bios['name'].str.startswith("James")), 
         ['name', 'born_country', 'born_date']].sort_values('born_date', ascending=False)

# Different way of doing this stuff, like shorthand
# Same as: bios.loc[(bios['born_country'] == "USA") & (bios['born_city'] == "Dallas")]
bios.query('born_country == "USA" and born_city == "Dallas"')
```
