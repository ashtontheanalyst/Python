## Merging and Concatenating Data

```python
import pandas as pd
```

```python
# bios is the olympics data
# NOC is National Olympics Committee
nocs = pd.read_csv('./data/noc_regions.csv')
bios = pd.read_excel('./data/olympics-data.xlsx')
nocs.info()
bios.info()
```

```python
# Merging Data between bios and nocs

bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC',
                    suffixes=["_bios", "_nocs"], how='left')
bios_new.rename(columns={'region' : 'born_country_full'}, inplace=True)

# Seeing who was born in another country but now under a different NOC, only show the three columns
bios_new[bios_new['NOC_bios'] != bios_new['born_country_full']][['name', 'NOC_bios', 'born_country_full']]

# See people only from the USA or GBR
usa = bios[bios['born_country'] == 'USA'].copy()
gbr = bios[bios['born_country'] == 'GBR'].copy()

# New data frame with those two combined into one
ug = pd.concat([usa, gbr])

ug.sample(10)
```
